```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
target = df.Survived
inputs = df.drop("Survived", axis = 'columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs, dummies], axis = 'columns')
inputs.head(3)
```

```python
inputs.drop('Sex', axis = 'columns', inplace = True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.2)
```

```python
print("X_train data types:")
print(X_train.dtypes)
print("\nFirst few rows:")
print(X_train.head())
print(f"\nX_train shape: {X_train.shape}")
```

```python
X_train = X_train.drop(['Name', 'Ticket', 'Cabin'], axis=1)
X_test = X_test.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
X_train.fillna({'Age': X_train['Age'].mean(), 'Embarked': X_train['Embarked'].mode()[0]}, inplace=True)
X_test.fillna({'Age': X_test['Age'].mean(), 'Embarked': X_test['Embarked'].mode()[0]}, inplace=True)

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X_train['Embarked'] = le.fit_transform(X_train['Embarked'])
X_test['Embarked'] = le.transform(X_test['Embarked'])
```

```python
print(X_train.dtypes)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
score = model.score(X_test, y_test)
print("Accuracy:", score)
```
