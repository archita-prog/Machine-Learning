```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("House Price India.csv")
df
```

```python
import math # Make sure math is imported

# Use bracket notation for the column name
median_bedrooms = math.floor(df['number of bedrooms'].median())
median_bedrooms
```

```python
df['number of bedrooms'].fillna(median_bedrooms)
df
```

```python
print(df.columns)
```

```python
import datetime
current_year = datetime.datetime.now().year
df['house_age'] = current_year - df['Built Year']
```

```python
reg = linear_model.LinearRegression()
# Use the renovation year if available, otherwise use the built year
df['age_since_update'] = current_year - df[['Built Year', 'Renovation Year']].max(axis=1)

# Now fit the model with these new features
reg.fit(df[['living area', 'number of bedrooms', 'age_since_update']], df.Price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
# 1. Get the coefficients and the intercept from the model
coefficients = reg.coef_
intercept = reg.intercept_

print(f"Coefficients (slopes): {coefficients}")
print(f"Intercept (base price): {intercept}")

# 2. Perform the full manual calculation
# These are the values you used for living area, bedrooms, and age
living_area = 3000
bedrooms = 3
age = 40

# The complete formula
manual_price = (coefficients[0] * living_area) + (coefficients[1] * bedrooms) + (coefficients[2] * age) + intercept

# 3. Get the model's prediction
model_prediction = reg.predict([[living_area, bedrooms, age]])[0]

# Now they will match!
print(f"\nModel Prediction:   {model_prediction}")
print(f"Manual Calculation: {manual_price}")
```
